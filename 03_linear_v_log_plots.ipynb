## **_Instructions:_**

-   Create two datasets: one consisting of values with a linear relationship, the other with logarathmic related values.

    -   The sets should span $0 < x <= 10$.

    -   $x$ cannot be $0$ since we're computing the logarithm.

-   Graph both sets twice, first on a plot with a linear-scaled y-axis, then on a plot with a logarithmic-scaled axis.


---

## **_My solution:_**

```python
# Import libraries:
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create set of x values:
x = np.linspace(0.01, 10, 100)
# First y set values same as x:
y_equal_x = x
# Second y set values e**x:
y_equal_e_x = np.exp(x)

# Create subplots with two axis columns:
fig, ax = plt.subplots(1, 2, figsize=(7.5, 4))

# Create label/title variables to avoid WET code:
y_x_label = "$y = x$"
y_e_x_label = "$y = e^x$"
titles = ["Linear y-axis", "Logarithmic y-axis"]

# Set logarithmic-scale for second subplot:
ax[1].set_yscale('log')

# Iterate over lin and log subplots to avoid repeated commands:
for i in range(len(ax)):
    # Plot linear values:
    ax[i].plot(x, y_equal_x,
               'dodgerblue', label=y_x_label, linewidth=2)
               
    # Plot logarithmic values:
    ax[i].plot(x, y_equal_e_x,
               'firebrick', label=y_e_x_label, linestyle='--', linewidth=2)
    ax[i].set_title(titles[i], fontdict={'fontweight': 'bold'})

    # Clarify bounds for x-axes shared by both plots:
    ax[i].set_xlim([0, 10])
    ax[i].set_box_aspect(1)
    ax[i].legend(loc='upper right')

# Adjust y-limits for better look at the plots:
ax[0].set_ylim([0, 100])
ax[1].set_ylim([0.1, 10**2])

# Shorten y-axis range for clarity of both y sets:
ax[1].set_yticks([10**i for i in range(-2, 3)])
# Turn of unnecessary minor ticks:
ax[1].minorticks_off()

# Adjust padding around subplots:
plt.tight_layout()
# Visualize:
plt.show()
```

---

## **_Conclusions:_**

-   On the linear y-axis plot, the logarithmic y-values appear as an exponentially growing curve, the linear y-values appear to appear as a straight line climbing at a fixed geometric rate.

-   On the logarithmic y-axis plot we have the opposite: the log y-values look to have a steady/proportional growth, the linear y-values appear to have an negative exponential decaying curve.

-   In other words, with a linear-scaled y-axis a linear function is a straight line and an logarithmic function is a quickly accelerating curve. On a log-scaled y-axis the linear function climbs at a rate which becomes steadily lower and the logarithmic function appears as a straight line.

-   For a simple set of x-values and y-values like this I think a linear plot is preferrable. I look forward to seeing more complex functions where an logarithmic y-axis is better suited.
