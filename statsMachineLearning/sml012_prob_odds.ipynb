## **_Background:_**

-   Exploring the relationship between the **PROBABILITY** formula and the **ODDS** ratio.

-   We learned the formula for probability based on `n` and `m`, the numbers in an odds ratio, _e.g., `n:m`.

<hr style="height: 0; box-shadow: 0 0 5px 4px dodgerblue; width: 80%;">

### **_The Probability Formula:_**

Probability can be calculated by taking the number of times a specific outcome occurs and dividing it by the total number of possible outcomes:

$$
\LARGE
    \begin{gathered}
        \text{probability of outcome }
        \textcolor{turquoise}{i}
        \\ = \\
        \frac{
                \textcolor{dodgerblue}{c} \text{ount of times outcome }
                \textcolor{turquoise}{i}
                \text{ can occur}
                }
            {
                \textcolor{dodgerblue}{c} \text{ount of all possible outcomes}
                }
    \end{gathered}
$$

$$
\Huge
    \Downarrow
$$    

$$
\huge
    p_{\textcolor{turquoise}{i}}
    =
    \frac{
            \textcolor{dodgerblue}{c}_{\textcolor{turquoise}{i}}}
            {\sum \textcolor{dodgerblue}{c}
        }
$$

<hr style="height: 0; box-shadow: 0 0 5px 4px dodgerblue; width: 80%;">

### **_Explaining the Odds Ratio:_**

$$
\LARGE
    \begin{gathered}
        \text{odds ratio }
        \textcolor{limegreen}{r}
        \\ = \\
        \frac{
                1 - \textcolor{gold}{p} \text{robability of specific outcome occuring}
                }
            {
                \textcolor{gold}{p} \text{robability of specific outcome occuring}
                }
    \end{gathered}
$$

$$
\Huge
    \Downarrow
$$    

$$
\huge
    \textcolor{limegreen}{r}
    \; = \;
    \frac{1 -  \textcolor{gold}{p}}{\textcolor{gold}{p}}
    \quad \text{ or } \quad
    \textcolor{limegreen}{r}
    \; = \;
    1 -  \textcolor{gold}{p} \; : \; \textcolor{gold}{p}
$$

$$
\Huge
    \Downarrow
$$    

We refer to `1 - p`, the chances of a specific outcome NOT occuring, as `n`.

We refer to `p`, the chances of a specific outcome ACTUALLY occuring, as `m`.

$$
\huge
    \textcolor{limegreen}{r}
    \; = \;
    \frac{\textcolor{violet}{n}}{\textcolor{fuchsia}{m}}
    \quad \text{ or } \quad
    \textcolor{limegreen}{r}
    \; = \;
    \textcolor{violet}{n} \; : \; \textcolor{fuchsia}{m}
$$

$$
\huge
    \textcolor{limegreen}{r}
    \; = \;
    \frac{1 -  \textcolor{gold}{p}}{\textcolor{gold}{p}}
    \quad \text{ or } \quad
    \textcolor{limegreen}{r}
    \; = \;
    1 -  \textcolor{gold}{p} \; : \; \textcolor{gold}{p}
$$

#### **_Solve for $\textcolor{gold}{p}$, in terms of $\textcolor{fuchsia}{m}$ and $\textcolor{violet}{n}$..._**

$$
\huge
    \textcolor{limegreen}{r}
    \; = \;
    \frac{1 -  \textcolor{gold}{p}}{\textcolor{gold}{p}}
$$

$$
\Huge
    \Downarrow
$$    

$$
\huge
    \textcolor{gold}{p}
        \; = \;
    \frac{1}{1 + \textcolor{limegreen}{r}}
$$

$$
\Huge
    \Downarrow
$$    

This is the probability formula we'll use in our code:

$$
\huge
    \textcolor{gold}{p}
    \; = \;
    \frac{1}{1 + \frac{\textcolor{violet}{n}}{\textcolor{fuchsia}{m}}}
$$

## **_Goals:_**

Create a two-dimensional matrix (image) of probability values with:

-   A range for `m`, the chances of an event occuring, on the x-axis.

-   A range for `n`, the chances of an event not occuring, on the y-axis.

-   Clarify results with a colorbar indicating range of probability values.

-   Compute the overall probabilities in terms of `m` and `n` with the formula we discerned:

$$
\huge
    \textcolor{gold}{p}
    \; = \;
    \frac{1}{1 + \frac{\textcolor{violet}{n}}{\textcolor{fuchsia}{m}}}
$$

-   Analyze, considering how `m` relates to `n`, and how this probability formula relates to the original odds ratio.

<hr style="height: 0; box-shadow: 0 0 5px 4px crimson; width: 95%;">

## **_Code and Analysis:_**

````python
import matplotlib.pyplot as plt
import numpy as np

m = 13
n = 37

p = np.zeros((m, n))
for i in range(m):
    for j in range(n):
        p[i, j] = 1 / (1 + (j + 1) / (i + 1))

plt.figure(figsize=(7, 11))

plt.imshow(p.T, cmap='Blues_r', origin='lower')
plt.xticks([i for i in range(0, m, m // 4)], [i for i in range(1, m + 1, m // 4)])
plt.yticks([i for i in range(0, n, n // 4)], [i for i in range(1, n + 1, n // 4)])

plt.xlabel("$m$")
plt.ylabel("$n$")

plt.title("$n$ (Probability AGAINST) vs. $m$ (Probability OF)\n")

plt.colorbar()
plt.show()
````

`n` is the probability of the event not occurring. `m` is the probability of the event occurring.

The odds ratio places `n` in the numerator. or to the left of the colon. `m` is in the denominator, or to right of the colon.

The probability is close to 0 when `m` is low and `n` is high.

When there is little probability of the event occurring (`m`) and high probability of the event not occurring (`n`), the probability is very low. In this case of low probability the odds ratio is higher (larger numerator, larger `n`).

Conversely, the probability is close to 1 (highest possible) when `m` is high and `n` is low. When there is high probability of the event actually occurring (`m`) and low probability of the event not occurring (`n`), the probability is very high. In this case of high probability the odds ratio is lower (larger denominator, larger `m`).

So in the above example where the odds are "Thirty-Seven to Thirteen Against", or `37 : 13`:

-   `37` represents `n`, the probability of the event _not_ occurring.

-   `13` is `m`, the probability of the event _actually_ occurring.

-   `n` is much higher than `m`, so the odds are fairly low.

If the odds were `13` : `37`, we'd have a significantly better chance of the event happening, whatever that event happens to be...

### **_Another Example:_**

````python
mn_total = 292201338
m = np.linspace(1, mn_total, 1000)
n = np.linspace(1, mn_total, 1000)

p = np.zeros((len(m), len(n)))
for i in range(len(m)):
    for j in range(len(n)):
        p[i, j] = 1 / (1 + ((n[j]) / (m[i])))

fig, ax = plt.subplots(figsize=(17, 11))

plt.imshow(p.T, cmap='plasma', origin='lower', extent=[m[0], m[-1], n[0], n[-1]], aspect='auto')

# Plot 1 ticket purchased:
tix_1 = 1
cost_1 = tix_1 * 3
ax.plot(
    tix_1,
    mn_total - tix_1,
    mfc='orangered',
    marker='o',
    ms=20,
    mec='lime',
    mew=1,
    label=f"Buy: {tix_1:,} ticket\nCost: ${cost_1:,.2f}\nOdds: {mn_total:,} : {tix_1:,}",
)

# Plot 10,000,000 tickets:
tix_10000000 = 10000000
cost_10000000 = tix_10000000 * 3
ax.plot(
    tix_10000000,
    mn_total - tix_10000000,
    mfc='orangered',
    marker='*',
    ms=17.5,
    mec='lime',
    mew=1,
    label=f"Buy: {tix_10000000:,} tickets\nCost: ${cost_10000000:,.2f}\nOdds: {mn_total:,} : {tix_10000000:,}",
)

# Plot 100,000,000 tickets:
tix_100000000 = 100000000
cost_100000000 = tix_100000000 * 3
ax.plot(
    tix_100000000,
    mn_total - tix_100000000,
    mfc='orangered',
    marker='X',
    ms=15,
    mec='lime',
    mew=1,
    label=f"Buy: {tix_100000000:,} tickets\nCost: ${cost_100000000:,.2f}\nOdds: {mn_total:,} : {tix_100000000:,}",
)

# Plot 250 million tickets:
tix = 250000000
cost_250000000 = tix * 3
ax.plot(
    tix,
    mn_total - tix,
    mfc='lime',
    marker='H',
    ms=15,
    mec='orangered',
    mew=2,
    label=f"Buy: {tix:,} tickets\nCost: ${cost_250000000:,.2f}\nOdds: {mn_total:,} : {tix:,}",
)

# Plot 250 million tickets:
tix_292201338 = 292201338
cost_292201338 = tix_292201338 * 3
ax.plot(
    tix_292201338,
    mn_total - tix_292201338,
    mfc='lime',
    marker='s',
    ms=20,
    mec='orangered',
    mew=2,
    label=f"Buy: {tix_292201338:,} tickets\nCost: ${cost_292201338:,.2f}\nOdds: {mn_total // tix_292201338} : {mn_total // tix_292201338}",
)

ax.set_xlabel("$m$ (Probability OF)")
ax.set_ylabel("$n$ (Probability AGAINST)")

ax.set_xticks(np.linspace(m[0], m[-1], num=5))
ax.set_yticks(np.linspace(n[0], n[-1], num=5))

title = f"Odds of Winning the ${20000000:,} Powerball Jackpot:\n"
title += f"{292201338:,} Possible Winning Combinations\n"

ax.set_title(title, size=24)
fig.legend(
    loc="outside lower center",
    bbox_to_anchor=(0.5, -0.1),
    fancybox=True,
    shadow=True,
    ncol=5,
    fontsize=12,
)

ax.grid(color='cyan', linestyle="--", alpha=0.35, lw=1.5)

plt.colorbar()
plt.tight_layout()
plt.show()
````

-   The Powerball draws five balls numbers `1` through `69`, and a bonus ball between `1` and `26`.

-   The total number of possible combinations is `292,201,338`, which represents `n` + `m`.

-   If we buy one ticket where we indicate five regular numbers and one bonus number, `m` will equal 1, since this is one instance of all the possible combinations of winners.

-   `n` is `292,201,337`, which is `1` subtracted from the total number of possible combinations. This represents all of the winning combinations we did not choose.

-   The odds are `292,201,337 : 1`, the marker located in dark purple in the visualization, tightly in the upper-left hand corner. Very poor odds.

-   What if we buy ten million tickets, each with a different series of numbers?

-   This would cost about **thirty million** US dollars. The jackpot starts at twenty million dollars, so this would be entirely irrational. But nevertheless...

-   The odds are `292,201,338 - 10,000,000 : 10,000,000`, or `282,201,338 : 10,000,000`.

-   This reduces the odds ratio to about `29:1`. We're situated much better, but still somewhere in the purple range, slightly closer to magenta, and still firmly in the upper-left hand corner of the plot.

-   We buy ten million tickets and still have only a one in twenty-nine chance of winning the jackpot.

-   If we do win the jackpot, we're still down ten million dollars.

-   Perhaps the only winning move is not to play?
