## **_Instructions:_**

-   The formula for converting a histogram from counts to proportion:

$$
\large
    \text{bin}_i = 100 \times \text{bin}_i / \text{sum}(\text{bins})
$$

-   And proportion:

$$
\large
    \text{bin}_i = \text{bin}_i / \text{sum}(\text{bins})
$$

-   In the previous lesson we plotted a histogram of the raw counts from a data set by extracting the values with `np.hist()` and creating a line-plot with `plt.plot()`.

-   Modify the previous lesson's code to convert the histogram from raw counts into percentage, then proportion, on the y-axis.

---

## **_My solution:_**

#### **_First things:_**

```python
# Import libraries:
import matplotlib.pyplot as plt
import numpy as np
```

#### **_Create variables:_**

```python
# Indicate number of points:
n = 1250

# Generate log-normal dataset:
data = np.exp(np.random.randn(n) / 2)

# Number of histogram bins:
k = 40

# Get values/bounds from np.histogram():
y_count, x = np.histogram(data, bins=k)
total_count = np.sum(y_count)

# Use the formula to convert raw count
#  to percentage:
y_perc = 100 * y_count / np.sum(y_count)
total_perc = np.sum(y_perc)

# Use the formula to convert raw count
#  to proportion:
y_prop = y_count / np.sum(y_count)
total_prop = np.sum(y_prop)

# Center the bars around bins' means:
xx = (x[1:] + x[:-1]) / 2
```

#### **_Plot:_**

```python
# Instantiate plot elements, adjust size manually:
fig, ax = plt.subplots(1, 3, figsize=(13, 5))

# Formula to normalize bin width:
bar_width = max(xx) / k

# Bar plot with edges, bin width variable:
ax[0].bar(
    xx,
    y_count,
    color='dodgerblue',
    edgecolor='k',
    width=bar_width,
    label=f"Count total: {total_count}"
)
ax[0].set_title("Data: Raw Count")
ax[0].set_xlabel("Bin Values")
ax[0].set_ylabel("Bin Size")
ax[0].legend()

ax[1].bar(
    xx,
    y_perc,
    color='yellowgreen',
    edgecolor='k',
    width=bar_width,
    label=f"Percentage total: {total_perc}"
)
ax[1].set_title("Data: Percentage")
ax[1].set_xlabel("Bin Values")
ax[1].set_ylabel("Bin Size")
ax[1].legend()

ax[2].bar(
    xx,
    y_prop,
    color='tomato',
    edgecolor='k',
    width=bar_width,
    label=f"Proportion total: {total_prop}"
)
ax[2].set_title("Data: Proportion")
ax[2].set_xlabel("Bin Values")
ax[2].set_ylabel("Bin Size")
ax[2].legend()

plt.suptitle("Histograms: Count and Proportion", fontsize=18)
plt.tight_layout()
plt.show()
```

---

## **_Conclusions:_**

1. I created subplots for easy comparison of the original plot with the new one. A histogram based on a dataset's raw count has the same shape as one based on proportion. They look the same at first glance.

2. The nature of the raw to proportion conversion formula necessary leaves us with a distribution that adds to $100$.

2. Though we're measuring the same data, we're looking at it in two different ways. Raw data is a good way to analyze a specific data set in isolation or alongside similarly sized datasets. Proportion is a better choice when comparing with disproportionately-sized datasets, or when we just want to understand the percentage of each sub-set/bin of values.

3. Bar-width is an issue when plotting histograms via `plt.plot()`, so I had to devise a formula to compensate for this variable: divide the maximum of the dataset by the number of bins. In the context of this Matplotlib function this creates well-spaced bars.
